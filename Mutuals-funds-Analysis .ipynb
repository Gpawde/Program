```python
# Calculator
def calculator(investment, r, years):
    months = years* 12

    #SIP_AMOUNT = float(input("Enter your sip amount "))
    i = r/100/12

    c= (1 + i)**months- 1

    b = (1 + i)/i

    fv = investment * c * b
    print(fv)
    return fv
```
